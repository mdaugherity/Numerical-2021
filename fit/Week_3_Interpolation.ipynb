```
xdata = np.arange(0,10)
ydata = np.sin(xdata)

f = interpolate.interp1d(xdata,ydata,kind='linear')
x = np.linspace(0,9,num=100)
y = f(x)   # interpolate x

plt.plot(xdata,ydata,'ko')
plt.plot(x,y,color='crimson')
plt.title('Interpolation Example')
plt.xlabel('x [m]')
plt.ylabel('Hamster Velocity [m/s]')
plt.show()
```

```
xdata = np.arange(0,10)
ydata = np.sin(xdata)

plt.figure(figsize=(15,5))  # plot size in inches

plt.subplot(1,3,1)  # 1 row, 3 cols, plot 1
f = interpolate.interp1d(xdata,ydata,kind='linear')
x = np.linspace(0,9,num=100)
y = f(x)   # interpolate x
plt.plot(xdata,ydata,'ko')
plt.plot(x,y,color='crimson')
plt.title('Interpolation Example')
plt.xlabel('x [m]')
plt.ylabel('Hamster Velocity [m/s]')

plt.subplot(1,3,2)  # 1 row, 3 cols, plot 2
f = interpolate.interp1d(xdata,ydata,kind='nearest')
x = np.linspace(0,9,num=100)
y = f(x)   # interpolate x
plt.plot(xdata,ydata,'ko')
plt.plot(x,y,color='crimson')
plt.title('NEAREST')
plt.xlabel('x [m]')
plt.ylabel('Hamster Velocity [m/s]')

plt.subplot(1,3,3)  # 1 row, 3 cols, plot 3
f = interpolate.interp1d(xdata,ydata,kind='cubic')
x = np.linspace(0,9,num=100)
y = f(x)   # interpolate x
plt.plot(xdata,ydata,'ko')
plt.plot(x,y,color='crimson')
plt.title('CUBIC')
plt.xlabel('x [m]')
plt.ylabel('Hamster Velocity [m/s]')

plt.tight_layout()
plt.show()
```

# FINALLY THE RIGHT WAY

```
xdata = np.arange(0,10)
ydata = np.sin(xdata)

plt.figure(figsize=(15,5))  # plot size in inches

kinds = ['linear','quadratic','cubic']
NUM = len(kinds)

for i in range(NUM):
  plt.subplot(1,NUM,i+1)  # 1 row, N cols
  f = interpolate.interp1d(xdata,ydata,kind=kinds[i])
  x = np.linspace(0,9,num=100)
  y = f(x)   # interpolate x
  plt.plot(xdata,ydata,'ko')
  plt.plot(x,y,color='crimson')
  plt.title(kinds[i])
  plt.xlabel('x [m]')
  plt.ylabel('Hamster Velocity [m/s]')

plt.tight_layout()
plt.show()
```

# FITTING EXAMPLE

Demo of polyfit and polyval

```
xdata = np.arange(0,5)
m = 2
b = 5
ydata = m*xdata + b 
p = np.polyfit(xdata,ydata,1)
print(p)

x = np.linspace(0,4)
y = np.polyval(p, x)

plt.plot(xdata,ydata,'bs')
plt.plot(x,y,'g--')
plt.show()
```

```
xdata = np.arange(0,5)
a = 5
b = -6
c = 2
ydata = a*xdata**2 + b*xdata + c 
p = np.polyfit(xdata,ydata,2)
print(p)

x = np.linspace(0,4)
y = np.polyval(p, x)

plt.plot(xdata,ydata,'bs')
plt.plot(x,y,'g--')
plt.show()
```
